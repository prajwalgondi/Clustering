```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('crime_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
##### Normalization
```

```python
df1=StandardScaler().fit_transform(df)
```

```python
df1
```

```python
### Dbscan
```

```python
df2=DBSCAN(eps=1,min_samples=4)
df2.fit(df1)
```

```python
df2.labels_
```

```python
#### clustering
```

```python
df['clusters']=df2.labels_
```

```python
df
```

```python
df.groupby('clusters').agg(['mean']).reset_index()
```

```python
### visulation 
```

```python
plt.scatter(df['clusters'],df['UrbanPop'],c=df2.labels_)
```
